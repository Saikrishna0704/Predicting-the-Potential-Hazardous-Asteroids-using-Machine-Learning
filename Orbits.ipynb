```python
import rebound
import pandas as pd
```

```python
data=pd.read_csv("/content/drive/MyDrive/NEO data/Data of 27_04_2021.csv")

ate=data[data["class"]=="ATE"]
apo=data[data["class"]=="APO"]
ieo=data[data["class"]=="IEO"]
amo=data[data["class"]=="AMO"]
```

# **Orbit plot**
Function to plot the orbits 

```python
def orbit_plot(e,a):
  sim = rebound.Simulation()
  sim.add(m=1)
  sim.add(m=1e-3, e=sum(e)/len(e)	, a=sum(a)/len(a)) #orbit
  sim.add(m=1e-3, e=0.01671022, a=1.00000011) #Earth 
  fig, ax = rebound.OrbitPlot(sim,slices=0, fancy=True, color=['cyan','lime'],
                            xlim=None, ylim=None,unitlabel="[AU]",lw=1.5, figsize=(7,7),
                            orbit_type='solid',plotparticles=[1,2],Narc=1000)
  fig.savefig("image.png")
  fig.show()
```

# **Apollos vs Earth**
(lime color orbit shows the Earthâ€™s orbit and cyan color shows Near Earth orbit)

```python
e=list(apo['e'])
a=list(apo['a'])
orbit_plot(e,a)
```

# **Amors vs Earth**

(lime color orbit shows the Earthâ€™s orbit and cyan color shows Near Earth orbit)

```python
e=list(amo['e'])
a=list(amo['a'])
orbit_plot(e,a)
```
